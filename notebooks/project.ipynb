### Computation of NDVI 

```python
NIR_wavelength = 850 # nm
Red_wavelength = 650  # nm

NIR = ds['reflectance'].sel(wavelengths=NIR_wavelength, method='nearest')
Red = ds['reflectance'].sel(wavelengths=Red_wavelength, method='nearest')

NDVI = (NIR - Red) / (NIR + Red)

ds['NDVI'] = NDVI
```

```python
colors = plt.cm.Blues_r(np.linspace(0, 1, 128))
colors = np.vstack((colors, plt.cm.YlGn(np.linspace(0, 1, 128))))
custom_colormap = mcolors.LinearSegmentedColormap.from_list('blue_green', colors)


y, x = np.where(ds['NDVI'].values > 0.8)

ds['NDVI'].plot(cmap=custom_colormap, vmin=-1, vmax=1)
plt.gca().invert_xaxis()
plt.title("Variation of NDVI across CapeTown Region")
plt.text(700, 450, 'Table Mountain \n National Park', fontsize = 7)
plt.text(900, 850, 'Newlands \n Forest', fontsize = 7)
```
